Variables de Decision

```python
# Crear el modelo
m = Model("FunnysCompany")

# Y[c,p]: 1 si se instala/opera una planta tipo p en la ciudad c, 0 e.o.
Y = m.addVars(ciudades, tipos_planta, vtype=GRB.BINARY, name="Y")

# Z[c,t]: 1 si la planta en la ciudad c decide usar el transporte t, 0 e.o.
Z = m.addVars(ciudades, transportes, vtype=GRB.BINARY, name="Z")

# X[c,r,t,a]: Cantidad de unidades enviadas desde ciudad c a región r, usando transporte t en el año a
X = m.addVars(ciudades, regiones, transportes, años, vtype=GRB.CONTINUOUS, name="X")

m.update()
```

# Funcion Objetivo

```python
#SUMAS

# 1. Costos de Apertura (Se paga 1 vez si se abre la planta)
costo_apertura_total = quicksum(costos_apertura[c, p] * Y[c, p] 
                                   for c in ciudades for p in tipos_planta)

# 2. Costos Fijos (Se pagan por 3 años si la planta opera)
costo_fijo_total = quicksum(3 * costos_fijos[c, p] * Y[c, p] 
                               for c in ciudades for p in tipos_planta)

# 3. Costos Variables de Producción (Dependen de las unidades fabricadas/enviadas)
costo_variable_total = quicksum(costos_variables[c, "Pequeña"] * X[c, r, t, a] 
                                   for c in ciudades for r in regiones 
                                   for t in transportes for a in años)

# 4. Costos de Transporte (Costo unitario de envío por la cantidad de unidades enviadas)
costo_transporte_total = quicksum(costos_transporte[t, c, r] * X[c, r, t, a] 
                                     for c in ciudades for r in regiones 
                                     for t in transportes for a in años)

# Funcion Objetivo
m.setObjective(costo_apertura_total + costo_fijo_total + 
               costo_variable_total + costo_transporte_total, GRB.MINIMIZE)
```

## Restricciones

```python
# 1. Satisfacer la demanda por región y por año
for r in regiones:
    for a in años:
        # Calcular demanda dinámica para el año 'a'
        demanda_proyectada = demanda_actual[r] * ((1 + tasas_crecimiento[r]) ** a)
        
        # Agregamos la restricción al modelo
        m.addConstr(
            quicksum(X[c, r, t, a] for c in ciudades for t in transportes) == demanda_proyectada,
            name=f"Demanda_{r}_Anio_{a}"
        )
# 2. Capacidad Máxima de Producción
for c in ciudades:
    for a in años:
        m.addConstr(
            quicksum(X[c, r, t, a] for r in regiones for t in transportes) <= 
            quicksum(capacidad_planta[p] * Y[c, p] for p in tipos_planta),
            name=f"Capacidad_{c}_Anio_{a}"
        )

# 3. Límite de Instalaciones por Ciudad
for c in ciudades:
    m.addConstr(
        quicksum(Y[c, p] for p in tipos_planta) <= 1,
        name=f"Max_Una_Planta_{c}"
    )
# 4. Transporte Único por Planta
for c in ciudades:
    m.addConstr(
        quicksum(Z[c, t] for t in transportes) <= 
        quicksum(Y[c, p] for p in tipos_planta),
        name=f"Un_Transporte_Maximo_{c}"
    )

# 5. Relación Lógica entre Envíos (X) y Transporte (Z) con "Big-M"
M_grande = capacidad_planta["Grande"]

for c in ciudades:
    for t in transportes:
        for a in años:
            m.addConstr(
                quicksum(X[c, r, t, a] for r in regiones) <= M_grande * Z[c, t],
                name=f"Vinculo_Envio_Transporte_{c}_{t}_Anio_{a}"
            )
# 6. Condición especial para Rancagua (Planta ya existe)
m.addConstr(
    Y["Rancagua", "Pequeña"] + Y["Rancagua", "Grande"] == 1,
    name="Planta_Existente_Rancagua"
)
```

## Resultados

```python
m.optimize()

#Print resultados 

if m.status == GRB.OPTIMAL:
    print(f"\n¡Solución óptima encontrada! Costo Total: ${m.objVal:,.0f}")
    
    print("\n--- Decisiones de Plantas (Y) y Transporte (Z) ---")
    for c in ciudades:
        for p in tipos_planta:
            if Y[c, p].x > 0.5:
                print(f"-> Se abre una planta {p} en {c}.")
                for t in transportes:
                    if Z[c, t].x > 0.5:
                        print(f"   Utilizará el medio de transporte: {t}")
else:
    print("\nEl modelo no encontró una solución óptima (puede ser infactible).")
```
